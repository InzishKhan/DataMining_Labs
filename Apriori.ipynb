```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules



# Convert all columns to string type and concatenate them to create unique items
for col in df.columns:
    df[col] = df[col].astype(str) + '_' + col

# Create a list of transactions
transactions = df.values.tolist()

# Perform Apriori algorithm
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)

# Generate rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Display rules
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
# Rank the top 10 association rules for different combinations of support and confidence
support_confidence_combinations = [(0.5, 0.7), (0.6, 0.8), (0.7, 0.9)]

for support, confidence in support_confidence_combinations:
    frequent_itemsets = apriori(df, min_support=support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=confidence)
    top_10_rules = rules.sort_values(by='confidence', ascending=False).head(10)
    print(f"Support={support}, Confidence={confidence}")
    print(top_10_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
    print()
```

```python
# Rank the top 10 association rules for different combinations of support and interest
support_interest_combinations = [(0.5, 0.5), (0.6, 0.6), (0.7, 0.7)]

for support, interest in support_interest_combinations:
    frequent_itemsets = apriori(df, min_support=support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=interest)
    top_10_rules = rules.sort_values(by='lift', ascending=False).head(10)
    print(f"Support={support}, Interest={interest}")
    print(top_10_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
    print()
```

```python
# Store rules in result variable
results = list(rules)# See the items that were bought together with their support
results_list = []
for i in range(0, len(results)):
    results_list.append('RULE:\t' + str(results[i][0]) + '\nSUPPORT:\t' + str(results[i][1]))
```
